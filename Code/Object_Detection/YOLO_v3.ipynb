```
%cd /content/darknet-master
#===============================================================================================================================
# -------------------{ Training : Execute the below command if running from previous saved weights }----------------------------
#===============================================================================================================================

# the first argument is the path of your yolo.data file ex: data/yolo.data
# the second is the path of yolov3_custom_train.cfg file ex: cfg/yolov3_custom_train.cfg
# the third is the path to your previous trained weights file ex: backup/yolov3_custom_train_2.backup
!./darknet detector train data/yolo.data cfg/yolov3_custom_train.cfg backup/yolov3_custom_train_2.backup
```

```
#================================================================================
# -------------------{ Image Testing Command }-----------------------------------------
#================================================================================
# the first argument is the path of your yolo.data file ex: data/yolo.data
# the second is the path of yolov3_custom_train.cfg file ex: cfg/yolov3_custom_train.cfg
# the third is the path to your previous trained weights file ex: backup/yolov3_custom_train_2.backup

!./darknet detector test data/yolo.data cfg/yolov3_custom_train.cfg backup/yolov3_custom_train_2.backup data/images/01.png
```

```
#================================================================================
# -------------------{ Video Testing Command }-----------------------------------------
#================================================================================
# the first argument is the path of your yolo.data file ex: data/yolo.data
# the second is the path of yolov3_custom_train.cfg file ex: cfg/yolov3_custom_train.cfg
# the third is the path to your previous trained weights file ex: backup/yolov3_custom_train.backup
# the dont_show flag prevents popping a video screen while processing the video (which may slow down the process)
# the fifth argument is the original video file
# the last argument is the name you will give the ouput processed video
!./darknet detector demo data/yolo.data cfg/yolov3_custom_train.cfg yolov3_custom_train_2.backup -dont_show assembled5.mp4 -i 0 -out_filename assembled5_out2.mp4
```
